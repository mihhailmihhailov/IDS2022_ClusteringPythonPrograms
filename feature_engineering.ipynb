```python
import pandas as pd
```

```python
df = pd.read_csv("df45.csv", encoding="utf-8", sep=";")
df.head(20)
```

```python
# how many lines were dedicated to a comment starting with a "#"
df["commented_lines"] = df["Tekst"].str.count("#[^#]")
# how many lines were left blank
df["blank_lines"] = df["Tekst"].str.count("\n\n")
# how many lines had code in them
df["code_lines"] = df["Tekst"].str.count("\n") - df["commented_lines"] - df["blank_lines"] + 1
# how many symbols were there on average per line
df["characters_per_line"] = df["Tekst"].str.len() / (df["commented_lines"] + df["blank_lines"] + df["code_lines"])
# how many times was a class defined
df["class_definitions"] = df["Tekst"].str.count("class .+ *:")
# how many times was a function defined
df["function_definitions"] = df["Tekst"].str.count("def .+\(.*\) *:")
# how many times was a function called (note that this will also count how many times objects were created because the syntax is the same)
df["function_calls"] = df["Tekst"].str.count(".+\(.*\)[^:]")
# how many times was a module from a library imported
df["imported_modules"] = df["Tekst"].str.count("from .+ import")
# how many times was an entire library imported
df["imported_libararies"] = df["Tekst"].str.count("import") - df["imported_modules"]
# how many times was an object's property accessed (with the syntax of "object.property")
df["properties_accessed"] = df["Tekst"].str.count("[^ ]\.[^ ]")
# how many times was a variable assigned a value
df["variable_assignments"] = df["Tekst"].str.count("[^=+-<>]=[^=]")
# how many times were values of variables compared with "=="
df["variable_comparisons"] = df["Tekst"].str.count("==")
# how many times were values of variables compared with "!="
df["variable_not_comparisons"] = df["Tekst"].str.count("!=")
# how many times was a value of a variable incremented with "+="
df["variable_increments"] = df["Tekst"].str.count("\+=")
# how many times was a value of a variable decremented with "-="
df["variable_decrements"] = df["Tekst"].str.count("-=")
# how many times was a piece of code written with camel-case syntax (format is "camelCase")
df["camel_case_instances"] = df["Tekst"].str.count("[a-zäöüõšž][A-ZÄÖÜÕŠŽ]")
# how many times was a piece of code written with snake-case syntax (format is "snake_case")
df["snake_case_instances"] = df["Tekst"].str.count("[a-zäöüõšž]_[a-zäöüõšž]|[A-ZÄÖÜÕŠŽ]_[A-ZÄÖÜÕŠŽ]")
# how many times was an array (or rather a list) declared
df["array_declarations"] = df["Tekst"].str.count("[ =]\[.*\]")
# how many times was a value of an array accessed by index
df["accesses_by_index"] = df["Tekst"].str.count("\[.*\]") - df["array_declarations"]
# how many times was a set declared
df["set_declarations"] = df["Tekst"].str.count("set\(.*\)")
# how many times was a string literal used (strings literals being text between "" or '')
df["string_literals"] = df["Tekst"].str.count("\".*\"|'.*'")
# how many times was an if statement used
df["if_statements"] = df["Tekst"].str.count("[\n| ]if .+:")
# how many times was an elif statement used
df["elif_statements"] = df["Tekst"].str.count("[\n| ]elif .+:")
# how many times was an else statement used
df["else_statements"] = df["Tekst"].str.count("[\n| ]else *:")
# how many times was a try statement used
df["try_statements"] = df["Tekst"].str.count("[\n| ]try *:")
# how many times was an except statement used
df["except_statements"] = df["Tekst"].str.count("[\n| ]except *:")
# how many times was a with statement used
df["with_statements"] = df["Tekst"].str.count("[\n| ]with .+:")
# how many times was a not statement used
df["not_statements"] = df["Tekst"].str.count(" not ")
# how many times was a while loop used
df["while_loops"] = df["Tekst"].str.count("[\n| ]while .+:")
# how many times was a for loop used
df["for_loops"] = df["Tekst"].str.count("[\n| ]for .+:")
df.head(20)
```

```python
df.to_csv("df_with_features.csv", encoding="utf-8", index=True, sep=";")
```
